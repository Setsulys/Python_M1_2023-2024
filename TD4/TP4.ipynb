##### server

```python
def today():
    today = datetime.datetime.today()
    return cli.DateTime(today)

server = serv.SimpleXMLRPCServer(("localhost",8050))
print("listening on port 8050 ...")
server.register_function(today,"today")
server.serve_forever()
```
